```python
%matplotlib notebook
import pandas as pd
from IPython.display import display, HTML
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import axes3d
import csv
import math
```

```python
x_0, y_0 = 532290, 1382690


x, y, z, u, v, w = [], [], [], [], [], []

with open('../data_files/n_constant.wind', newline='') as csvfile:
    wind_level = csv.reader(csvfile, delimiter=' ', quotechar='|')
    for row in wind_level:
        x.append(x_0)
        y.append(y_0)
        z.append(float(row[0]))
        u.append(float(row[1]) * np.cos(math.radians(float(row[2])) + np.pi/2))
        v.append(float(row[1]) * np.sin(math.radians(float(row[2])) + np.pi/2))
        w.append(0)
    
fig = plt.figure()
ax = fig.gca(projection='3d')

ax.quiver(x, y, z, u, v, w, length=1000, normalize=False)

plt.show()
```
