```python
%matplotlib notebook 
import numpy as np
import matplotlib.pyplot as plt
```

```python
vent_easting, vent_northing = (532290, 5651176) # Vent coordinates in Easting Northing (UTM coordinates)

km_west = 50    # Number of km west of volcano to consider
km_east = 100   # Number of km east of volcano to consider
km_north = 100  # Number of km north of volcano to consider
km_south = 50   # Number of km south of volcano to consider

elevation = 350 # Elevation of landing surface

spacing = 1.5   # Spacing between square grid points (km)
```

```python
x_0 = vent_easting - km_west*1000
x_f = vent_easting + km_west*1000
y_0 = vent_northing - km_west*1000
y_f = vent_northing + km_north*1000

nx = int(np.round((x_f - x_0)/(spacing*1000)))
ny = int(np.round((y_f - y_0)/(spacing*1000)))

x = np.linspace(x_0, x_f, nx)
y = np.linspace(y_0, y_f, ny)

xx, yy = np.round(np.meshgrid(x, y))
```

```python
plt.scatter(xx, yy, s=1, c='k', marker='.')
plt.plot(vent_easting, vent_northing, 'ro')
```

```python
f = open('ruapehu_grid_file.pos', 'w')

for i, c in enumerate(xx):
    for j, p in enumerate(c):
        f.write(str(int(p)) + " " + str(int(yy[i][j])) + " " +  str(elevation) + "\n")
f.close()
```
