```python
%matplotlib notebook
import pandas as pd
from IPython.display import display, HTML
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import axes3d
from matplotlib.colors import LogNorm
import csv
import math

```

```python
sources = [
    '../cerroNegro_regGrid_constN_SOURCE.txt',
    '../cerroNegro_regGrid_constN_ONLINE.txt',
    '../cerroNegro_regGrid_constNE_SOURCE.txt',
    '../cerroNegro_regGrid_constNE_ONLINE.txt',
    '../cerroNegro_regGrid_constESE_SOURCE.txt',
    '../cerroNegro_regGrid_constESE_ONLINE.txt',
    '../cerroNegro_regGrid_noWind_SOURCE.txt',
    '../cerroNegro_regGrid_noWind_ONLINE.txt',
]

for s in sources:
    df = pd.read_csv(s ,sep=" ", header=0)
    df = df.dropna(axis=1, how='all')
    df = df.fillna(0)
    df = df.rename(columns={'#Easting': 'Easting'})
    xx = df['Easting'].values
    yy = df['Northing'].values
    zz_ma = df['Mass/Area'].values
    
    x_0, y_0 = 532290, 1382690 #Vent Location

    zz_ma = zz_ma + 0.000000000000000000001 #Adding epsilon to enable log plot
    
    plt.figure()
    plt.ylabel("Northing")
    plt.xlabel("Easting")
    contour = plt.tricontourf(xx, yy, zz_ma, 20, cmap='gnuplot2', norm=LogNorm(vmin=zz_ma.min(), vmax=zz_ma.max()));
    cbar = plt.colorbar()
    cbar.ax.set_ylabel('Mass/Area', rotation=270)
    plt.axis('equal')
    plt.plot(xx, yy, 'ko', ms=0.5)
    plt.plot(x_0, y_0, 'k*', ms=10)
    plt.tight_layout()
    plt.show()
    
    
```
