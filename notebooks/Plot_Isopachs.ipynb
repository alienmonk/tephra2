```python
%matplotlib notebook
import pandas as pd
from IPython.display import display, HTML
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import axes3d
from matplotlib.colors import LogNorm
import csv
import math

```

```python

df = pd.read_csv('../cerroNegro_regGrid_noWind_SOURCE.txt',sep=" ", header=0)


df = df.dropna(axis=1, how='all')
df = df.rename(columns={'#Easting': 'Easting'})
display(df)
```

```python
xx = df['Easting'].values
yy = df['Northing'].values
zz_ma = df['Mass/Area'].values
x_0, y_0 = 532290, 1382690

zz_ma = zz_ma + 0.000000000000000000001

print(zz_ma.min())

plt.figure()
plt.plot(xx, yy, 'ko', ms=1)
plt.plot(x_0, y_0, 'r*', ms=10)
plt.show()
```

```python
plt.figure()
plt.ylabel("Northing")
plt.xlabel("Easting")

contour = plt.tricontourf(xx, yy, zz_ma, 20, cmap='gnuplot2', norm=LogNorm(vmin=zz_ma.min(), vmax=zz_ma.max()));
cbar = plt.colorbar()
cbar.ax.set_ylabel('Mass/Area', rotation=270)
plt.axis('equal')
# plt.axis([520000, 540000, 1380000, 1405000])
# plt.clabel(contour, inline=1, fontsize=10)
plt.plot(xx, yy, 'ko', ms=0.5)
plt.plot(x_0, y_0, 'k*', ms=10)

plt.tight_layout()
plt.show()
```
