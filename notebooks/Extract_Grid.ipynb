```python
%matplotlib notebook
import pandas as pd
from IPython.display import display, HTML
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import axes3d
import csv
import math
```

```python
df = pd.read_csv('../cerroNegro_regGrid_constN_ONLINE.txt',sep=" ", header=0)


df = df.dropna(axis=1, how='all')
df = df.rename(columns={'#Easting': 'Easting'})
display(df)
```

```python
xx = df['Easting'].values
yy = df['Northing'].values
elev = df['Elev.'].values

plt.figure()
plt.plot(xx, yy, 'ko', ms=1)
plt.show()
```

```python
f = open('../data_files/cerroNegro_onlineStandard.pos', 'w')

for x, y, z in zip(xx, yy, elev):
    print(x, y, z)
    f.write(str(int(x)) + " " + str(int(y)) + " " +  str(int(z)) + "\n")
f.close()
```
