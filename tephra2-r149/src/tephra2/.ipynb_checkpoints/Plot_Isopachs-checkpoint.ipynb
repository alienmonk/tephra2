```python
%matplotlib notebook
import pandas as pd
from IPython.display import display, HTML
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d

```

```python

df = pd.read_csv('CerroNegroOutput.txt',sep=" ", header=0)


df = df.dropna(axis=1, how='all')
df = df.rename(columns={'#Easting': 'Easting'})
display(df)
```

```python
xx = df['Easting'].values
yy = df['Northing'].values
zz_ma = df['Mass/Area'].values

plt.figure()
plt.plot(xx, yy, 'ko', ms=1)
plt.show()
```

```python
plt.figure()
plt.ylim(ymax=1390000)
plt.ylim(ymin=1360000)
plt.xlim(xmax=540000)
plt.xlim(xmin=480000)
plt.tricontourf(xx, yy, zz_ma, 20, cmap='RdGy');
plt.colorbar()
plt.plot(xx, yy, 'ko', ms=1)
plt.show()
```

```python
plt.figure()
plt.ylim(ymax=1390000)
plt.ylim(ymin=1360000)
plt.xlim(xmax=540000)
plt.xlim(xmin=480000)
ax = plt.axes(projection='3d')
ax.plot_trisurf(xx, yy, zz_ma,
                cmap='viridis', edgecolor='none');
plt.show()
```
